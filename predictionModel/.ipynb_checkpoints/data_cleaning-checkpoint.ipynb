```python
#function to clean the data
def convertStringtoInt(df,lkey):
    labels=np.zeros(df[lkey].values.size,dtype=int)

    label_str=df[lkey].values
    for tt,tkey in enumerate(df.groupby(lkey).groups.keys()):
        ind=label_str==tkey
        labels[ind]=tt
        print(tkey,tt)
        
    return labels   


def cleanData(df):
    #keep the rows with atleast 17 non NA values 
    df.dropna(thresh = 17, inplace = True)
    df.reset_index(drop=True, inplace=True)
    
    #replacing null values in the dataframe 
    df['Applicant_Qualification'].fillna('Others', inplace = True)
    df['Applicant_Occupation'].fillna('Others', inplace = True) 
    df['Applicant_City_PIN'].fillna(0, inplace = True)
    
    df['Manager_Num_Application'] = df['Manager_Num_Application'].apply(np.int64)
    df['Manager_Num_Coded'] = df['Manager_Num_Coded'].apply(np.int64)
    df['Manager_Num_Products'] = df['Manager_Num_Products'].apply(np.int64)
    df['Manager_Num_Products2'] = df['Manager_Num_Products2'].apply(np.int64)
    
    
    ##converting the city pin from float to int fot computation and foramtting
    df['Applicant_City_PIN'] = df['Applicant_City_PIN'].apply(np.int64)
    
    list_column=['Applicant_Qualification','Manager_Current_Designation','Manager_Joining_Designation',
                'Applicant_Marital_Status','Applicant_Gender','Applicant_Occupation',
                'Manager_Status', 'Manager_Gender']
    
    for ll, lkey in enumerate(list_column):
        print('*** converting %s ****'%(lkey))
        df[lkey] = convertStringtoInt(df,lkey)
    
    
    #calculating age of manager and applicants at the time of joining.
    df["Applicant_BirthDate"] = pd.to_datetime(df["Applicant_BirthDate"])
    df["Application_Receipt_Date"] = pd.to_datetime(df["Application_Receipt_Date"])
    df["Manager_DOJ"] = pd.to_datetime(df["Manager_DOJ"])
    df["Manager_DoB"] = pd.to_datetime(df["Manager_DoB"])
    
    #adding new columns for the applicant and manager age at the time of joining
    df["Applicants_joining_age"] = df["Application_Receipt_Date"].dt.year - df["Applicant_BirthDate"].dt.year
    df["Manager_joining_age"] = df["Manager_DOJ"].dt.year - df["Manager_DoB"].dt.year
    
    #Replacing null values in applicants age with the average age of applicant grouped by occupation#
    g = df.groupby('Applicant_Occupation').Applicants_joining_age.mean()
    Occupation = g.index.values.tolist()
    Avg_age = g.values
    age_str = df['Applicants_joining_age']

    for index, rows in df.iterrows():
        for i in range(0,len(Occupation)):
            if rows['Applicant_Occupation'] == Occupation[i] and np.isnan(rows['Applicants_joining_age']) :
                age_str[index] = Avg_age[i]

    df['Applicants_joining_age'] = age_str
    
    return df
    
training_data = cleanData(df_train)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
#calculating the split size and splitting the index in k folds
#should be executed only once as it shuffles the data in random order in every execution 
class kfoldValidation:
    def __init__(self, datasize,k):
        self.k = k
        self.foldsize=np.int(np.floor(datasize*1.0/k))
        self.index_arr=np.arange(datasize)
        np.random.shuffle(self.index_arr)
        
    def getfold(self,x):
        if(self.k-1==x):
            index_test=self.index_arr[x*self.foldsize:]
            index_train=self.index_arr[:x*self.foldsize]
        else:
            index_test=self.index_arr[x*self.foldsize:((x+1)*self.foldsize)]
            index_train=np.append(self.index_arr[:x*self.foldsize],self.index_arr[(x+1)*self.foldsize:])

        return index_test, index_train
    
    
```

```python
#this class is performing the training and validation on data

def select_data_label(indata):
    data_list=[]
    reject_list=['Business_Sourced','ID','Office_PIN','Application_Receipt_Date',
                 'Applicant_BirthDate','Manager_DOJ','Manager_DoB',
                 #'Applicant_City_PIN','Manager_Grade','Manager_Business','Manager_Business2',
                #'Applicants_joining_age'
                ]
    for tkey in indata.columns:
        if(tkey not in reject_list):
            data_list.append(tkey)
            
    return indata[data_list],indata['Business_Sourced'].values


def perform_kfold_validation(train_data,train_labels,kernel='linear',k=10):
    
    #kfoldValidation needs to be executed only once as it shuffles the data in every execution and updates the index
    kfold = kfoldValidation(train_labels.size,k)

    for ii in range(k):
        
        print('working on fold (with kernel= %s):%d/%d'%(kernel,ii,k))
        
        #to get the test and training data index in each fold
        test_index,train_index=kfold.getfold(ii)

        #training the model on train data for each fold
        train_vectors = train_data.loc[train_index]
        test_vectors = train_data.loc[test_index]

        #SVM RBF/Gaussian Kernel (default)
        classifier = svm.SVC(kernel=kernel)
        classifier.fit(train_vectors, train_labels[train_index]) #train classifier on train data
        prediction = classifier.predict(test_vectors) #predict test data

        #get classification results/evaluate model
        report = metrics.precision_recall_fscore_support(train_labels[test_index], prediction)
        report_accuracy = accuracy_score(train_labels[test_index], prediction)
        auc=metrics.roc_auc_score(train_labels[test_index], prediction)
        
        if(ii==0):
            precision=report[0]
            recall=report[1]
            f1=report[2]
            accuracy=np.array([report_accuracy])
            auc_arr=np.array([auc])
        else:
            precision=np.row_stack([precision,report[0]])
            recall=np.row_stack([recall,report[1]])
            f1=np.row_stack([f1,report[2]])
            accuracy=np.append(accuracy,report_accuracy)     
            auc_arr=np.append(auc_arr,auc)
            
    return precision, recall,f1,accuracy, auc_arr

#calling the above function and taking the value of k as 10
k=10
train,train_lab=select_data_label(training_data)

for kernel in ['linear']:#,'poly','sigmoid','rbf']:
    precision, recall,f1,accuracy=perform_kfold_validation(train,train_lab,kernel=kernel,k=k)
    
    print('kernel used: ',kernel)
    print('mean overall accuracy =',np.mean(accuracy))
    print('mean precision= ',np.mean(precision),'mean precision per class',np.mean(precision,axis=0))
    print('mean recall   = ',np.mean(recall),'mean recall per class',np.mean(recall,axis=0))
    print('mean f1       = ',np.mean(f1),'mean f1 per class',np.mean(f1,axis=0))
    print('mean auc      = ',np.mean(auc_arr), np.std(auc_arr))
    
print('Finished')
```
